```python
# import all functions / classes from the tkinter
from tkinter import *
from textblob import TextBlob
import tkinter as tk
import re

file_ob = open('big.txt')

def spell_check(input):
    first_char = input[0]
    sort_input = ''
    extra = ''
    extra_sort = ''
    replace = ''
    recommend = []
    for i in range(1, len(input)):
        sort_input = sort_input+input[i]

    sort_input = ''.join(sorted(sort_input))
    sort_input = first_char+sort_input
    recommend.append(find_miss(sort_input))

    for i in range(1, len(input)):
        extra = input[:i]+input[i+1:]
        recommend.append(find_extra_char(extra))

    for i in range(1, len(input)):
        extra = input[:i]+input[i+1:]
        extra_sort = extra[1:]
        recommend.append(find_miss(extra_sort))

    for i in range(1, len(input)):
        integer = 96+i
        replace = input[1:i]+chr(integer)+input[i+1:]
        recommend.append(find_wrong_place(replace))

    sort_input = input[1:]
    sort_input = ''.join(sorted(sort_input))
    sort_input = first_char+sort_input
    recommend.append(find_replace(sort_input))

    return recommend


def find_miss(input):
    output = ''
    for line in file_ob.readlines():
        output = line
        if(line[0] == input[0] and len(line) == len(input)):
            found = check_miss(output, input)
            if(found == 1):
                return line


def check_miss(output, input):
    first_char = output[0]
    sort_output = ''
    for i in range(1, len(output)):
        sort_output = sort_output+input[i]
    sort_output = ''.join(sorted(sort_output))
    sort_output = first_char+sort_output
    if(sort_output == input):
        print("output= ", sort_output)
        return 1


def find_extra_char(input):
    for line in file_ob.readlines():
        if(line[0] == input[0] and len(line) == len(input)):
            if(line == input):
                return line


def find_wrong_place(input):
    for line in file_ob.readlines():
        if(line[0] == input[0] and len(line) == len(input)):
            if(line == input):
                return line


def correction():

    # get a content from entry box
    input_word = word1_field.get()

    # create a TextBlob object
    blob_obj = TextBlob(input_word)

    # get a corrected word
    corrected_word = str(blob_obj.correct())

    # insert method inserting the
    # value in the text entry box.
    word2_field.insert(10, corrected_word)


def find_replace(input):
    output = ''
    for line in file_ob.readlines():
        output = line
        if(line[0] == input[0] and len(line) == len(input)):
            found = check_replace(output, input)
            if(found == 1):
                return line


def check_replace(output, input):
    first_char = output[0]
    sort_output = ''
    for i in range(1, len(output)):
        sort_output = sort_output+input[i]
    sort_output = ''.join(sorted(sort_output))
    sort_output = first_char+sort_output
    if(sort_output == input):
        print("output= ", sort_output)
        return 1
# Function to clear both the text entry boxes


def clearAll():

    # whole content of text entry area is deleted
    word1_field.delete(0, END)
    word2_field.delete(0, END)


# Driver code
if __name__ == "__main__":

    # Create a GUI window
    root = Tk()

    # Set the background colour of GUI window
    root.configure(background='black')

    # Set the configuration of GUI window (WidthxHeight)
    root.geometry("650x315")

    # set the name of tkinter GUI window
    root.title("Spell Corrector")

    # Create Welcome to Spell Corrector Application: label
    headlabel = Label(root, text='Welcome to Spell Corrector Application',
                      fg='white', bg="black", font=('Times', 20))

    # Create a "Input Word": label
    label1 = Label(root, text="Input Word",
                   fg='black', bg='sienna2', font=('Helvetica bold', 10))

    # Create a "Corrected Word": label
    label2 = Label(root, text="Corrected Word",
                   fg='black', bg='sienna1', font=('Helvetica bold', 10))

    # grid method is used for placing
    # the widgets at respective positions
    # in table like structure .
    # padx keyword argument used to set padding along x-axis .
    headlabel.grid(row=0, column=1)
    label1.grid(row=1, column=0)
    label2.grid(row=3, column=0, padx=10)

    # Create a text entry box
    # for filling or typing the information.
    word1_field = Entry(width=30, justify='center', bd=0,
                        bg='light grey', font='sans-serif')
    word2_field = Entry(width=30, justify='center', bd=0,
                        bg='light grey', font='sans-serif')

    # padx keyword argument used to set padding along x-axis .
    # pady keyword argument used to set padding along y-axis .
    word1_field.grid(row=1, column=1, padx=20, pady=20)
    word2_field.grid(row=3, column=1, padx=20, pady=20)

    # Create a Correction Button and attached
    # with correction function
    button1 = Button(root, text="Check my Word", bg="sienna2", fg="black",
                     command=correction, font=('Helvetica bold', 10))

    button1.grid(row=2, column=1)

    # Create a Clear Button and attached
    # with clearAll function
    button2 = Button(root, text="Clear Input & Output", bg="sienna1",
                     fg="black", command=clearAll, font=('Helvetica bold', 10))

    button2.grid(row=4, column=1)

    # Start the GUI
    root.mainloop()
```
